### Resizing IMAGES

```python
datarute = r'C:\Users\Maqui\Documents\INTERNSHIP\Temporales\UTKFace\UTKFace'
new_datarute = r'C:\Users\Maqui\Documents\INTERNSHIP\CNN\UTKFace-crop_Renamed'


# Create a new folder to store the renamed images
os.makedirs(new_datarute, exist_ok=True)


for filename in os.listdir(datarute):
    # Split the filename into parts
    temp = filename.split('_')
    age = temp[0]
    gender = temp[1]
    temp2 = temp[-1].split('.')
    name = temp2[0]
    image_extension = temp2[-1]

    # Open the image using PIL
    image_path = os.path.join(datarute, filename)
    image = Image.open(image_path)

    # Crop 4 pixels from all sides
    cropped_image = image.crop((4, 4, image.width - 4, image.height - 4))

    # Create the new filename using the age, gender, and counter
    new_filename = f'{age}_{gender}_{name}.{image_extension}'

    # Create the full path for the new file
    new_path = os.path.join(new_datarute, new_filename)

    # Save the cropped image with the new filename
    cropped_image.save(new_path)
```

### Eliminando bits

```python
import matplotlib.pyplot as plt
import numpy as np

# image path
image_path = r"C:\Users\Maqui\Desktop\52_0_20170104212228222.jpg"


image = Image.open(image_path)
image = image.convert('L')
image = np.array(image)

rows, cols = image.shape
image_248 = np.zeros((rows,cols))
print(rows, cols)


for i in range(rows):
    for j in range(cols):
        image_248[i,j] = image[i,j]&248


fig, axes = plt.subplots(1, 2, figsize=(16, 8))
axes[0].imshow(image, cmap="gray")
axes[0].set_title("Original")
axes[1].imshow(image_248, cmap="gray")
axes[1].set_title("Imagen sin bits menos significativos")

plt.tight_layout()
plt.show()



new_datarute = r'C:\Users\Maqui\Documents\INTERNSHIP\CNN\Nuevo_Dataset'

# Create a new folder to store the renamed images
os.makedirs(new_datarute, exist_ok=True)

# Create the full path for the new file
new_filename = "image_bit248.png"
new_path = os.path.join(new_datarute, new_filename)

image = Image.fromarray(image.astype('uint8'))
image.save(os.path.join(new_datarute, "Original.jpeg"))

# De np.array a PIL.Image
image_248 = Image.fromarray(image_248.astype('uint8'))
image_248.save(new_path)
```
